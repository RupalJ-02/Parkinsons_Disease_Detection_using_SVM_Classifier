# **PARKINSON DISEASE DETECTION AND PREDICTION USING DEEP LEARNING**
Parkinson's disease is a progressive brain disorder that causes movement problems, mental health issues, and other health problems. It's a neurodegenerative disease that affects the central nervous system.

**Objective**: This project aims to develop a machine learning model for predicting Parkinson's disease using voice measurements.

**Importance**: Early detection of Parkinson's disease is vital for effective treatment and management.

**Approach**: We use a dataset of biomedical voice measurements, train an SVM model, and evaluate its performance.

**Impact**: The model can assist medical professionals in diagnosing Parkinson's disease and improving patient care.

# Attribute Information:

Matrix column entries (attributes):

name - ASCII subject name and recording number

MDVP:Fo(Hz) - Average vocal fundamental frequency

MDVP:Fhi(Hz) - Maximum vocal fundamental frequency

MDVP:Flo(Hz) - Minimum vocal fundamental frequency

MDVP:Jitter(%),MDVP:Jitter(Abs),MDVP:RAP,MDVP:PPQ,Jitter:DDP - Several
measures of variation in fundamental frequency

MDVP:Shimmer,MDVP:Shimmer(dB),Shimmer:APQ3,Shimmer:APQ5,MDVP:APQ,Shimmer:DDA - Several measures of variation in amplitude

NHR,HNR - Two measures of ratio of noise to tonal components in the voice

status - Health status of the subject (one) - Parkinson's, (zero) - healthy
RPDE,D2 - Two nonlinear dynamical complexity measures

DFA - Signal fractal scaling exponent

spread1,spread2,PPE - Three nonlinear measures of fundamental frequency variation

# Data Pre-Processing

## Seperating Other variables and Target Variables

```python
X = parkinson_data.drop(columns=['status','name'], axis=1) # Note : dropping column axis = 1; dropping row then axis = 0
Y = parkinson_data['status']
```

```python
print(X)
```

```python
print(Y)
```

## Splitting data into testing and training

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

# Model Training

### We will train the model using Support Vector Machine (SVM)

```python
model = svm.SVC(kernel='linear')
```

```python
# training the SVM model with training data
model.fit(X_train, Y_train)
```

# Model Evaluation

```python
# Accuracy Score on training data
X_train_pred = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_pred)
```

```python
print('Accuracy (Training Data) :', training_data_accuracy*100, '%')
```

```python
# Accuracy Score on test data
X_test_pred = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test, X_test_pred)
```

```python
print('Accuracy (Testing Data) :', testing_data_accuracy*100, '%')
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, balanced_accuracy_score, roc_auc_score, average_precision_score

# 1. Predictions on Training Data
X_train_pred = model.predict(X_train)

# 2. KPI Calculations (using Y_train and X_train_pred)
y_probs_train = model.decision_function(X_train)

# Calculate KPIs
accuracy = accuracy_score(Y_train, X_train_pred)
precision = precision_score(Y_train, X_train_pred)
recall = recall_score(Y_train, X_train_pred)

# Confusion Matrix for Specificity
tn, fp, fn, tp = confusion_matrix(Y_train, X_train_pred).ravel()
specificity = tn / (tn + fp)

f1 = f1_score(Y_train, X_train_pred)

# other performance metrics
balanced_accuracy = balanced_accuracy_score(Y_train, X_train_pred)
roc_auc = roc_auc_score(Y_train, y_probs_train)  # Use y_probs_train from above
average_precision = average_precision_score(Y_train, y_probs_train)

# 3. Print KPIs
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"Specificity: {specificity}")
print(f"F1-Score: {f1}")
print(f"Balanced Accuracy: {balanced_accuracy}")
print(f"ROC AUC: {roc_auc}")
print(f"Average Precision: {average_precision}")
```
